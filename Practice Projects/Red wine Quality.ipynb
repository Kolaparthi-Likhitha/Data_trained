###### Data Processing

###### Analysing target variable

- In this dataset quality range is between 3 and 8
- We will divide quality range into two parts:
- good - 7 or greater than 7
- not good - less than 7
- performing data binning to classify and categorize the data

```python
df['quality']=df.quality.apply(lambda x:1 if x>=7 else 0)
```

```python
df['quality'].value_counts()
```

- performing feature selection methods
- Here we are using 'univariate selection method' to select the features that have strongest relationship with target variable

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from pandas import DataFrame
```

```python
X=df.drop('quality',axis=1)
y=df.quality
```

```python
bestfeatures=SelectKBest(score_func=chi2,k=7)  # finding 7 best features
fit=bestfeatures.fit(X,y)
```

```python
df_scores=pd.DataFrame(fit.scores_)
df_columns=pd.DataFrame(X.columns)
```

- concatenating data frames

```python
feature_scores=pd.concat([df_columns,df_scores],axis=1)
feature_scores.columns=['Feature_name','Score']
```

```python
feature_scores
```

```python
print(feature_scores.nlargest(7,'Score'))
```

###### Building model with the top 7 features

```python
new_X=df[['total sulfur dioxide','free sulfur dioxide','alcohol','citric acid','fixed acidity','volatile acidity','residual sugar']]
```

```python
new_X.head()
```

```python
new_X.shape
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(new_X, y, test_size = 0.2, random_state = 42)  
#here trained the model with only 7(new_X) features
```

###### Classification Models

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# write one function and call as many times as to check accuracy_score for different models

# classify function
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.metrics import accuracy_score
def metric_score(clf,X_train,X_test,y_train,y_test,train=True):
    
    if train:
        y_pred=clf.predict(X_train)
        print('--Train Result--')
        print(f'Accuracy score:{accuracy_score(y_train,y_pred)*100:.2f}%')
    
    elif train==False:
        y_pred=clf.predict(X_test)
        print('--Test Result--')
        print(f'Accuracy score:{accuracy_score(y_test,y_pred)*100:.2f}%')
        
    #cross-validation
    score = cross_val_score(clf, new_X, y, cv=5)
    print("CV Score:", np.mean(score)*100)
        
    
    
        
```

###### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
metric_score(lr,X_train,X_test,y_train,y_test,train=True) #for training score
metric_score(lr,X_train,X_test,y_train,y_test,train=False) #for testing score
```

###### Knn model

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier()
```

```python
knn.fit(X_train,y_train)
```

```python
metric_score(knn,X_train,X_test,y_train,y_test,train=True) #for training score
metric_score(knn,X_train,X_test,y_train,y_test,train=False) #for testing score
```

###### Decision Tree model

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt= DecisionTreeClassifier()
```

```python
dt.fit(X_train,y_train)
```

```python
metric_score(dt,X_train,X_test,y_train,y_test,train=True) #for training score
metric_score(dt,X_train,X_test,y_train,y_test,train=False) #for testing score
```

###### RandomForest model

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier()
```

```python
rf.fit(X_train,y_train)
```

```python
metric_score(rf,X_train,X_test,y_train,y_test,train=True) #for training score
metric_score(rf,X_train,X_test,y_train,y_test,train=False) #for testing score
```

###### support vector machine model

```python
from sklearn.svm import SVC
```

```python
sv=SVC()
```

```python
sv.fit(X_train,y_train)
```

```python
metric_score(sv,X_train,X_test,y_train,y_test,train=True) #for training score
metric_score(sv,X_train,X_test,y_train,y_test,train=False) #for testing score
```

By training the models, the testing accuracy score is highest in Random Forest model. So, we choose that model as the best fit model.

###### ROC curve

```python
pred_prob1=lr.predict_proba(X_test)
pred_prob2=knn.predict_proba(X_test)
```

```python
from sklearn.metrics import roc_curve
fpr1,tpr1,thresh1=roc_curve(y_test,pred_prob1[:,1],pos_label=1)
fpr2,tpr2,thresh2=roc_curve(y_test,pred_prob2[:,1],pos_label=1)
random_probs=[0 for i in range(len(y_test))]
p_fpr,p_tpr,_=roc_curve(y_test,random_probs,pos_label=1)
```

```python
from sklearn.metrics import roc_auc_score
auc_score1=roc_auc_score(y_test,pred_prob1[:,1])
auc_score2=roc_auc_score(y_test,pred_prob2[:,1])
print(auc_score1,auc_score2)
```

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn')

plt.plot(fpr1,tpr1,linestyle='--',color='red',label='LogisticRegression')
plt.plot(fpr2,tpr2,linestyle='--',color='green',label='KNN')
plt.plot(p_fpr,p_tpr,linestyle='--',color='blue')

plt.title("ROC curve")
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')

plt.legend(loc='best')
plt.savefig('ROC',dpi=300)
plt.show()
```

It is evident from the plot that AUC for Logistic Regression curve is bit higher than that for the KNN ROC curve. Therefore we say that logistic regression did a better job of classifying the quality of redwine in the dataset.
