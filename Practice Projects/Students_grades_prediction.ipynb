##### Treating the outliers

```python
#Creating a function to drop the outliers

def drop_outliers(df,col_name):
    iqr=1.5+(np.percentile(df[col_name],75)-np.percentile(df[col_name],25))
    df.drop(df[df[col_name]>(iqr+np.percentile(df[col_name],75))].index,inplace=True)
    df.drop(df[df[col_name]<(np.percentile(df[col_name],25)-iqr)].index,inplace=True)
```

#### Building the models

```python
x = df_f.drop(['CGPA','Seat No.'], axis = 1)
y = df_f['CGPA']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

### #LINEAR REGRESSION

```python
#LINEAR REGRESSION
from sklearn.linear_model import LinearRegression

from sklearn.metrics import mean_squared_error, r2_score,accuracy_score
lr= LinearRegression()
lr.fit(x_train, y_train)
y_pred = lr.predict(x_test)

# Scores:
print("Model training score:",lr.score(x_train,y_train)*100,'%')
print('Model Testing Score:',lr.score(x_test, y_test)*100,'%')
print('R2 Score:',r2_score(y_test, y_pred))
print('MSE:',mean_squared_error(y_test, y_pred))
print("RMSE: ",mean_squared_error(y_test, y_pred, squared=False))
```

### #RANDOM FOREST REGRESSOR

```python
#RANDOM FOREST REGRESSOR
 
from sklearn.ensemble import RandomForestRegressor
rfr= RandomForestRegressor()
rfr.fit(x_train, y_train)
y_pred = rfr.predict(x_test)

# Scores:
print("Model training score:",rfr.score(x_train,y_train)*100,'%')
print('Model Testing Score:',rfr.score(x_test, y_test)*100,'%')
print('R2 Score:',r2_score(y_test, y_pred))
print('MSE:',mean_squared_error(y_test, y_pred))
print("RMSE: ",mean_squared_error(y_test, y_pred, squared=False))
```

### #K NEIGHBOR REGRESSOR

```python
#K NEIGHBOR REGRESSOR
from sklearn.neighbors import KNeighborsRegressor
KNN= KNeighborsRegressor()
KNN.fit(x_train, y_train)
y_pred = KNN.predict(x_test)

# Scores:
print("Model training score:",KNN.score(x_train,y_train)*100,'%')
print('Model Testing Score:',KNN.score(x_test, y_test)*100,'%')
print('R2 Score:',r2_score(y_test, y_pred))
print('MSE:',mean_squared_error(y_test, y_pred))
print("RMSE: ",mean_squared_error(y_test, y_pred, squared=False))
```

### #RIDGE REGRESSION

```python
#RIDGE REGRESSION
from sklearn.linear_model import Ridge
ri=Ridge()
ri.fit(x_train, y_train)
y_pred = ri.predict(x_test)

# Scores:
print("Model training score:",ri.score(x_train,y_train)*100,'%')
print('Model Testing Score:',ri.score(x_test, y_test)*100,'%')
print('R2 Score:',r2_score(y_test, y_pred))
print('MSE:',mean_squared_error(y_test, y_pred))
```

### #DECISION TREE REGRESSOR

```python
#DECISION TREE REGRESSOR
from sklearn.tree import DecisionTreeRegressor
dtr=DecisionTreeRegressor(random_state=1)
dtr.fit(x_train, y_train)
y_pred = dtr.predict(x_test)

# Scores:
print("Model training score:",dtr.score(x_train,y_train)*100,'%')
print('Model Testing Score:',dtr.score(x_test, y_test)*100,'%')
print('R2 Score:',r2_score(y_test, y_pred))
print('MSE:',mean_squared_error(y_test, y_pred))
print("RMSE: ",mean_squared_error(y_test, y_pred, squared=False))
```

#### Determining Feature Importance values

```python
importances = dtr.feature_importances_

# Sort the feature importance in descending order
#
sorted_indices = np.argsort(importances)[::-1]
 
feat_labels = df_f.columns[1:]
 
for f in range(x_train.shape[1]):
    print("%2d) %-*s %f" % (f + 1, 30,
                            feat_labels[sorted_indices[f]],
                            importances[sorted_indices[f]]))
```

### #SUPPORT VECTOR REGRESSOR

```python
#SUPPORT VECTOR REGRESSOR
from sklearn.svm import SVR
svr= SVR()
svr.fit(x_train, y_train)
y_pred = svr.predict(x_test)

# Scores:
print("Model training score:",svr.score(x_train,y_train)*100,'%')
print('Model Testing Score:',svr.score(x_test, y_test)*100,'%')
print('R2 Score:',r2_score(y_test, y_pred))
print('MSE:',mean_squared_error(y_test, y_pred))
print("RMSE: ",mean_squared_error(y_test, y_pred, squared=False))
```

- As we have more features i.e., 43 columns. So, I have used dimensionality reduction algorithm 'Principal Component Analysis(PCA)' to solve this issue.

##### Applying PCA

```python
from sklearn.decomposition import PCA
pca=PCA()
x_train_pca=pca.fit_transform(x_train)
x_test_pca=pca.transform(x_test)
```

```python
explained_variance=pca.explained_variance_ratio_
```

```python
explained_variance
```

- It can be seen that first principal component is responsible for 48% variance. Similarly first 21 components are collectively having 85.3% in the feature set.
- Lets use first 21 component to train our algorithm 

- PCA reduces dimensions by taking linear combination of different features so that we can capture maximum variance

```python
from sklearn.decomposition import PCA

pca=PCA(n_components=21)
x_train_pca=pca.fit_transform(x_train)
x_test_pca=pca.transform(x_test)
```

#### Training the model and making predictions

```python
#LINEAR REGRESSION
lr_p=LinearRegression()
lr_p.fit(x_train_pca,y_train)

y_pred=lr_p.predict(x_test_pca)
```

```python
print('Linear Regression')
print("Model training score:",lr_p.score(x_train_pca,y_train)*100,'%')
print('Model Testing Score:',lr_p.score(x_test_pca, y_test)*100,'%')
print('R2 Score:',r2_score(y_test, y_pred))
print('MSE:',mean_squared_error(y_test, y_pred))
print("RMSE: ",mean_squared_error(y_test, y_pred, squared=False))
```

- After applying PCA, we find that accuracy of linear regression model has increased from 98.34% to 98.72%. There is a mild increase in the accuracy. So, PCA fit for this model.

```python
#RANDOM FOREST REGRESSOR
from sklearn.ensemble import RandomForestRegressor
rf_p=RandomForestRegressor(max_depth=2,random_state=0)
rf_p.fit(x_train_pca,y_train)

y_pred=rf_p.predict(x_test_pca)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error, r2_score,accuracy_score

print('Random Forest')
print("Model training score:",rf_p.score(x_train_pca,y_train)*100,'%')
print('Model Testing Score:',rf_p.score(x_test_pca, y_test)*100,'%')
print('R2 Score:',r2_score(y_test, y_pred))
print('MSE:',mean_squared_error(y_test, y_pred))
print("RMSE: ",mean_squared_error(y_test, y_pred, squared=False))
```

- After applying PCA, we find that accuracy of random forest model has reduced from 94.72% to 92.41%. So, PCA doesnt fit for this model.


```python
#K NEIGHBOR REGRESSOR
knn_p=KNeighborsRegressor()
knn_p.fit(x_train_pca,y_train)

y_pred=knn_p.predict(x_test_pca)
```

```python
print('K Neighbor Regressor')
print("Model training score:",knn_p.score(x_train_pca,y_train)*100,'%')
print('Model Testing Score:',knn_p.score(x_test_pca, y_test)*100,'%')
print('R2 Score:',r2_score(y_test, y_pred))
print('MSE:',mean_squared_error(y_test, y_pred))
print("RMSE: ",mean_squared_error(y_test, y_pred, squared=False))
```

- After applying PCA, we find that accuracy of k neighbor regression model has increased from 95.31% to 96.25%. So, PCA fit for this model.

```python
#Decision Tree Regressor
dtr_p=DecisionTreeRegressor(random_state=1)
dtr_p.fit(x_train_pca,y_train)

y_pred=dtr_p.predict(x_test_pca)
```

```python
print('Decision Tree Regressor')
print("Model training score:",dtr_p.score(x_train_pca,y_train)*100,'%')
print('Model Testing Score:',dtr_p.score(x_test_pca, y_test)*100,'%')
print('R2 Score:',r2_score(y_test, y_pred))
print('MSE:',mean_squared_error(y_test, y_pred))
print("RMSE: ",mean_squared_error(y_test, y_pred, squared=False))
```

- After applying PCA, we find that accuracy of decision tree regression model has increased from 78.30% to 96.63%. So, PCA fit for this model.

```python
#Ridge REGRESSOR
ri_p=Ridge()
ri_p.fit(x_train_pca,y_train)

y_pred=ri_p.predict(x_test_pca)
```

```python
print('Ridge Regressor')
print("Model training score:",ri_p.score(x_train_pca,y_train)*100,'%')
print('Model Testing Score:',ri_p.score(x_test_pca, y_test)*100,'%')
print('R2 Score:',r2_score(y_test, y_pred))
print('MSE:',mean_squared_error(y_test, y_pred))
print("RMSE: ",mean_squared_error(y_test, y_pred, squared=False))
```
